```python
def annotate_point(xy,  
                   text, 
                   xytext, 
                   ax,
                   relpos=(0.5, 0.5),
                   lw=0.5,
                   color='0.15',
                   fontsize=6,
                   fontweight='regular',
                   fontcolor='0.1',
                   ha='center',
                   zorder=6,
                   circle_size=20, 
                   bbox_pad=0,
                   bbox_facecolor='white',
                   highlight=False,
                   **kwargs
                  ):
    """
    Convenience function for streamlining annotation 
    on slope vs. slope and volcano plots. 
    
    Position of text is given in the difference in 'axes fraction' 
    away from the point for easier eyeballing. 
    
    xytext values should be a percentage (e.g. 8) rather than a decimal (0.08) 
    
    """
    frac_to_data = ax.transLimits.inverted().transform  # converts axes fraction to data coords
    # To calculate the xytext as a delta, find the difference in data coords from (0, 0)
    xtextdelta, ytextdelta = (frac_to_data(xytext) - frac_to_data([0, 0])) / 100
    xytext = (xy[0] + xtextdelta, xy[1] + ytextdelta)
    if highlight:
        bbox_pad = HIGHLIGHT_BBOX_PAD
        fontsize = HIGHLIGHT_FONTSIZE
        fontweight = HIGHLIGHT_FONTWEIGHT
        bbox_facecolor = HIGHLIGHT_FACECOLOR
        lw = HIGHLIGHT_ANNOT_LW
        zorder = zorder + 1
    ax.annotate(
        text=text, 
        xy=xy, 
        xytext=xytext, #(counter+5, row['coef']+0.05),
        textcoords='data',
        arrowprops=dict(arrowstyle='-', 
                        relpos=relpos, 
                        lw=lw, 
                        color=color),
        bbox=dict(pad=bbox_pad, 
                  facecolor=bbox_facecolor, 
                  edgecolor='none'),
        fontsize=fontsize, 
        fontweight=fontweight,
        color=fontcolor,
        annotation_clip=True, 
        ha=ha, va='center',
        zorder=zorder,
        **kwargs
    )
    # Draw circle around point
    alpha_corrected_color = (str(float(color) - 0.15))
    ax.scatter(xy[0], xy[1], 
               edgecolor=alpha_corrected_color, facecolor='white', linewidth=lw, alpha=0.7,
               s=circle_size, zorder=zorder)

round_to_n = lambda x, n: round(x, -int(floor(log10(x))) + (n - 1))

def find_nth(haystack, needle, n):
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
```

```python
def year_mapper(x, factor=8):
    if x >= 1500:
        return 1500 + (x-1500) * factor
    else:
        return x
    
def human_format(num):
    num = float('{:.2g}'.format(num))
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    return '{}{}'.format('{:f}'.format(num).rstrip('0').rstrip('.'), ['', 'K', 'M', 'B', 'T'][magnitude])
```

```python
fig, ax = plt.subplots(dpi=250, figsize=(3, 2))
df = data.iloc[:16].copy()
df['yearmap'] = df['End'].apply(lambda x: year_mapper(x))
df['Start'] = df['Start'].astype('int')


xcol = 'yearmap'
ycol = 'log'

ax.scatter(data=df, x=xcol, y=ycol, s=(df[ycol]**4)/50 - 20, c='darkred', lw=0.7, ec='0.0', zorder=10,)

ax.set_xlabel('Conflict end year')
ax.set_ylabel(None)
ax.set_title('Conflict death toll', loc='left', fontsize=8, pad=-2)
x_keep = [0, 500, 1000, 1500, 1600, 1700, 1800, 1900, 2000]
xs = range(0, 2100, 100)
xticklabels = [str(x) if (x in x_keep) else '' for x in x_keep]
# xticklabels[0] = '250 BCE'
ax.set_xticks([year_mapper(x) for x in x_keep], xticklabels, fontsize=5, rotation=90)
ax.set_xticks([100, 200, 300, 400, 600, 700, 800, 900, 1100, 1200, 1300, 1400], labels=['']*12, minor=True)
yticks = np.log10([2e6, 3e6, 5e6, 1e7, 2e7, 3e7, 5e7, 1e8])
ax.set_yticks(yticks, ['2M', '3M', '5M', '10M', '20M', '30M', '50M', '100M'], ha='left', va='bottom')

ax.set_ylim(bottom=df[ycol].min() - 0.2, top=8.2)
ax.set_axisbelow(True)
ax.set_xlim(left=-400)
ax.grid(axis='y', lw=0.5, color='0.8')
ax.tick_params(axis='y', length=0, pad=0)
ax.tick_params(axis='x', length=2.5, pad=1, rotation=0)
sns.despine(left=True)

offsets = {
    0:  dict(x=0, y=6, fontsize=7),  # ww2
    1:  dict(x=0, y=11.5, text='Mongol\nconquests', fontsize=7, ha='center', relpos=(0.5, 0.5)),  # mongol
    2:  dict(x=0, y=9, text='Taiping\nrebellion', fontsize=5, bbox_pad=-1),   # taiping
    3:  dict(x=0, y=15, text='European\ncolonization\nof Americas', fontsize=5),  # euro col
    4:  dict(x=-5, y=0, text='Ming to Qing\ntransition', fontsize=5, ha='right'), # ming to qing
    5:  dict(x=2, y=10, text='2nd Sino-\nJapanese war', ha='left', fontsize=5, relpos=(0,0)),  # 2nd sino-japanese
    6:  dict(x=-3.5, y=-4, fontsize=5, relpos=(1,1), ha='right'),   # ww1
    7:  dict(x=2.5, y=0, text='An Lushan\nrebellion', ha='left', fontsize=5, relpos=(1,0)),  # an-lushan
    8:  dict(x=-4, y=0, text='Dungan\nrebellion', fontsize=5, ha='right'),
    9:  dict(x=9, y=1, text='Chinese\ncivil war', fontsize=5),
    10: dict(x=9, y=-2, text='Russian\ncivil war', fontsize=5),
    11: dict(x=-4, y=3, text='Thirty years war', ha='right', fontsize=5, relpos=(1, 0)),
    12: dict(x=-4, y=-6.5, text='Mughal-Maratha war', fontsize=5),
    13: dict(x=0, y=-8.5, text='Napoleonic wars', fontsize=5, bbox_pad=-0.5),
    14: dict(x=5, y=0, text='Yellow\nturban\nrebellion', ha='left', fontsize=5, relpos=(0, 0.5)),
    15: dict(x=-11, y=-8, text='Second Congo war', fontsize=5, relpos=(0.8,1)),
}
for i, d in offsets.items():
    if 'text' in d:
        text = d.pop('text')
    else:
        text = df.iloc[i].loc['Event']
    xy = (df.iloc[i].loc['yearmap'], df.iloc[i].loc['log'])
    xytext = (d.pop('x'), d.pop('y'))
    annotate_point(xy=xy, text=text, xytext=xytext, ax=ax, **d)
plt.savefig('Top16_conflicts_death_toll.png', dpi=1000, bbox_inches='tight', facecolor='white')
```
