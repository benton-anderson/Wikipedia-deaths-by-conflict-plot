```python
def annotate_point(xy,  
                   text, 
                   xytext, 
                   ax,
                   relpos=(0.5, 0.5),
                   lw=0.5,
                   color='0.15',
                   fontsize=6,
                   fontweight='regular',
                   fontcolor='0.1',
                   ha='center',
                   zorder=6,
                   circle_size=20, 
                   bbox_pad=0,
                   bbox_facecolor='white',
                   highlight=False,
                   **kwargs
                  ):
    """
    Convenience function for streamlining annotation 
    on slope vs. slope and volcano plots. 
    
    Position of text is given in the difference in 'axes fraction' 
    away from the point for easier eyeballing. 
    
    xytext values should be a percentage (e.g. 8) rather than a decimal (0.08) 
    
    """
    frac_to_data = ax.transLimits.inverted().transform  # converts axes fraction to data coords
    # To calculate the xytext as a delta, find the difference in data coords from (0, 0)
    xtextdelta, ytextdelta = (frac_to_data(xytext) - frac_to_data([0, 0])) / 100
    xytext = (xy[0] + xtextdelta, xy[1] + ytextdelta)
    if highlight:
        bbox_pad = HIGHLIGHT_BBOX_PAD
        fontsize = HIGHLIGHT_FONTSIZE
        fontweight = HIGHLIGHT_FONTWEIGHT
        bbox_facecolor = HIGHLIGHT_FACECOLOR
        lw = HIGHLIGHT_ANNOT_LW
        zorder = zorder + 1
    ax.annotate(
        text=text, 
        xy=xy, 
        xytext=xytext, #(counter+5, row['coef']+0.05),
        textcoords='data',
        arrowprops=dict(arrowstyle='-', 
                        relpos=relpos, 
                        lw=lw, 
                        color=color),
        bbox=dict(pad=bbox_pad, 
                  facecolor=bbox_facecolor, 
                  edgecolor='none'),
        fontsize=fontsize, 
        fontweight=fontweight,
        color=fontcolor,
        annotation_clip=True, 
        ha=ha, va='center',
        zorder=zorder,
        **kwargs
    )
    # Draw circle around point
    alpha_corrected_color = (str(float(color) - 0.15))
    ax.scatter(xy[0], xy[1], 
               edgecolor=alpha_corrected_color, facecolor='white', linewidth=lw, alpha=0.7,
               s=circle_size, zorder=zorder)

    
round_to_n = lambda x, n: round(x, -int(floor(log10(x))) + (n - 1))


def find_nth(haystack, needle, n):
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
```

```python
def year_mapper(x):
    if x >= 1500:
        return 1500 + (x-1500) * 6.5
    else:
        return x
```

```python
fig, ax = plt.subplots(dpi=250, figsize=(3, 2))
df = data.iloc[:20].copy()
df['yearmap'] = df['End'].apply(year_mapper)

xcol = 'yearmap'
ycol = 'log'

sns.scatterplot(
    data=df, x=xcol, y=ycol, 
#     size='geomean', sizes=(10, 100),
    linewidth=0,
    zorder=10,
)
ax.set_xlabel('Conflict end year')
ax.set_ylabel('Conflict death toll')
xs = [-250, 0, 500, 1000, 1500, 1600, 1700, 1800, 1900, 2000]
xticklabels = [str(x) for x in xs]
xticklabels[0] = '250 BCE'
ax.set_xticks([year_mapper(x) for x in xs], xticklabels, fontsize=5, rotation=90)

ax.set_ylim(top=8.2)

ax.set_axisbelow(True)
# ax.get_yaxis().get_major_formatter().labelOnlyBase = False
# ax.set_yscale('log', )
# ax.grid()
# ax.set_
# ax.set_yticks([1e6, 2e6, 3e6, 4e6, 5e6, 6e6, 7e6, 8e6, 9e6, 10e6, 20e6])
# ax.set_yticks([1e6, 1e7, 1e8], ['hi', 'hi', 'hi'])
# ax.get_yaxis().set_major_formatter(plt.matplotlib.ticker.ScalarFormatter())

# ax.legend().remove()
sns.despine()

offsets = {
    0: (-15, 0),  # ww2
    1: (-10, 5, 'Mongol\nconquests'),  # mongol
    2: (10, 10, 'Taiping\nrebellion'),   # taiping
    3: (0, 15),  # euro col
    4: (-15, 10), # ming to qing
    5: (15, 15, '2nd Sino-\nJapanese war'),  # 2nd sino-japanese
    6: (15, 0),   # ww1
    7: (-11, 11, 'An Lushan\nrebellion'),  # an-lushan
    8: (-11, 8, 'Dungan\nrebellion'),
    9: (12, 0, 'Chinese\ncivil war'),
    10: (12, 0, 'Russian\ncivil war')
    
}
for i, row in df.iterrows():
    x = row[xcol]
    y = row[ycol]
    name = row['Altname']
    if len(name) > 10:
        pos = find_nth(name, ' ', 3)
#         if pos == -1:
#             pos = find_nth(name, ' ', 2)
        if pos != -1:
            name = list(name)
            name[pos] = '\n'
            name = ''.join(name)
    if offsets.get(i) and len(offsets[i]) > 2:
        name = offsets[i][2]
    text = name + f'\n{int(round_to_n(10**y/1e6, 2))}M'
    if i in offsets:
        xytext = (offsets[i][0], offsets[i][1])
    else:
        xytext = (4, 4)
    annotate_point(xy=(x, y), text=text, xytext=xytext, ax=ax,
        circle_size=2, fontsize=4.5,
        bbox_facecolor='none', bbox_pad=0
                  )
```
